```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd

dotenv.load_dotenv(override=True)

username=os.getenv("username")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)

query = """SELECT * FROM dimproduct"""

df = pd.read_sql(query, db_engine)
db_engine.dispose()
df
```

```python
#In media quanto pesano i prodotti?
peso_medio = df["Weight"].mean()
round(peso_medio, 2)
```

```python
#Quanto pesa il più leggero?
peso_minimo = df["Weight"].min()
peso_minimo
```

```python
#Quanto pesa il più pesante?
peso_massimo = df["Weight"].max()
peso_massimo
```

```python
#Quanti prodotti pesano più di 100 Kg?
peso_maggiore_100 = df[df["Weight"] > 100].shape[0]
peso_maggiore_100
```

```python
#Quanto costano in media i prodotti (colonna DealerPrice)?
prezzo_medio = df["DealerPrice"].mean()
round(prezzo_medio, 2)
```

```python
#Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?
percentile75 = df["DealerPrice"].quantile(0.75)

maggiore_percentile75 = df[df["DealerPrice"] >= percentile75]

min = maggiore_percentile75["DealerPrice"].min()
max = maggiore_percentile75["DealerPrice"].max()

percentile75, min, max
```

```python
#Qual è il prezzo medio per i prodotti di colore blu?
prodotti_blu = df[df['Color'] == 'Blue']
prezzo_medio_blu = prodotti_blu['DealerPrice'].mean()
round(prezzo_medio_blu, 2)
```

```python
#Qual è il prezzo medio per i prodotti di colore rosso o nero?
prodotti_neri_rossi = df[(df['Color'] == 'Red') | (df['Color'] == 'Black')]
prezzo_medio_neri_rossi = prodotti_neri_rossi['DealerPrice'].mean()
round(prezzo_medio_neri_rossi, 2)
```

```python
#Qual è il prezzo massimo per i prodotti di taglia 42 e peso oltre i 10 Kg?
prodotti_taglia42_peso10 = df[(df['Size'] == '42') & (df['Weight'] > 10)]
prezzo_max = prodotti_taglia42_peso10['DealerPrice'].max()
round(prezzo_max, 2)
```
